```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df = pd.read_csv('all_prices_extracted_cleaned_categorized_analyzed.csv')

def gerar_wordcloud(texto, titulo):
    wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=500).generate(texto)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation = 'bilinear')
    plt.axis('off')  # Remove os eixos
    plt.title(titulo, fontsize=18)
    plt.show()

for site in df['site'].unique():
    categorias = df[df['site'] == site]['categoria'].dropna().astype(str).unique()
    marcas = df[df['site'] == site]['brand'].dropna().astype(str).unique()
    modelos = df[df['site'] == site]['model'].dropna().astype(str).unique()
    submodelos = df[df['site'] == site]['brand_model'].dropna().astype(str).unique()
    cores = df[df['site'] == site]['color'].dropna().astype(str).unique()

    texto = ' '.join(set(categorias) | set(marcas) | set(modelos) | set(submodelos) | set(cores))
    
    gerar_wordcloud(texto, f'Word Cloud - Site: {site}')
```
